## Imports

Vamos começar com nossas importações. Aqui estamos importando o TensorFlow e chamando-o de tf para facilitar o uso.

Em seguida, importamos uma biblioteca chamada numpy, que nos ajuda a representar nossos dados como listas de maneira fácil e rápida.

A estrutura para definir uma rede neural como um conjunto de camadas sequenciais é chamada de keras, portanto, também a importamos.

````
import tensorflow as tf
import numpy as np
from tensorflow import keras
````

## Definir e compilar a rede neural

Em seguida, criaremos a rede neural mais simples possível. Tem 1 camada, e essa camada tem 1 neurônio, e a forma de entrada para ela é apenas 1 valor.

````
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
````

Agora compilamos nossa Rede Neural. Ao fazer isso, temos que especificar 2 funções, uma perda e um otimizador.

Se você já viu muita matemática para aprendizado de máquina, é aqui que geralmente é usado, mas, neste caso, está bem encapsulado em funções para você. Mas o que acontece aqui -- vamos explicar...

Sabemos que em nossa função, a relação entre os números é y=3x+1.

Quando o computador está tentando 'aprender' isso, ele faz uma suposição...talvez y=10x+10. A função PERDA mede as respostas adivinhadas em relação às respostas corretas conhecidas e mede o quão bem ou mal ele foi.

Em seguida, ele usa a função OPTIMIZER para fazer outro palpite. Com base em como foi a função de perda, ele tentará minimizar a perda. Nesse ponto, talvez apareça algo como y=5x+5, que, embora ainda seja muito ruim, está mais próximo do resultado correto (ou seja, a perda é menor)

Ele repetirá isso pelo número de EPOCHS que você verá em breve. Mas primeiro, aqui está como dizemos para usar 'MEAN SQUARED ERROR' para a perda e 'STOCHASTIC GRADIENT DESCENT' para o otimizador. Você não precisa entender a matemática para isso ainda, mas pode ver que eles funcionam! :)

Com o tempo, você aprenderá as diferentes e apropriadas funções de perda e otimizador para diferentes cenários.

````
model.compile(optimizer='sgd', loss='mean_squared_error')
````

# Fornecendo os dados

Em seguida, alimentaremos alguns dados. Neste caso, estamos tomando 6 xs e 6ys. Você pode ver que a relação entre eles é y=2x-1, então onde x = -1, y=-3 etc. etc.

Uma biblioteca python chamada 'Numpy' fornece muitas estruturas de dados do tipo array que são uma maneira padrão de fazer isso. Declaramos que queremos usá-los especificando os valores como np.array[]

````
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
````

# Treinando a Rede Neural

O processo de treinamento da rede neural, onde ela 'aprende' a relação entre os Xs e os Ys, está na chamada **model.fit**. É aqui que ele vai passar pelo loop que falamos acima, fazendo um palpite, medindo o quão bom ou ruim é (também conhecido como perda), usando o opimizer para fazer outro palpite etc. especificamos. Ao executar este código, você verá a perda no lado direito.

````
model.fit(xs, ys, epochs=500)
````

Ok, agora você tem um modelo que foi treinado para aprender a relação entre X e Y. Você pode usar o método **model.predict** para que ele descubra o Y para um X desconhecido anteriormente. Então, por exemplo, se X = 10, quanto você acha que Y será? Dê um palpite antes de executar este código:

````
print(model.predict([10.0]))
````

Você deve ter pensado 31, certo? Mas acabou ficando um pouco para trás. Por que você acha que é isso?

Lembre-se de que as redes neurais lidam com probabilidades, portanto, dados os dados com os quais alimentamos a NN, ela calculou que há uma probabilidade muito alta de que a relação entre X e Y seja Y=3X+1, mas com apenas 6 pontos de dados podemos não sei ao certo. Como resultado, o resultado para 10 é muito próximo de 31, mas não necessariamente 31.

Ao trabalhar com redes neurais, você verá esse padrão se repetir. Você quase sempre lidará com probabilidades, não certezas, e fará um pouco de codificação para descobrir qual é o resultado com base nas probabilidades, principalmente quando se trata de classificação.
